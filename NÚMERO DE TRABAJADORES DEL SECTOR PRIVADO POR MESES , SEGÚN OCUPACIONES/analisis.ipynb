```python
df.columns = ['COD_OCUPACIONES', 'OCUPACIONES', 'ENERO', 'FEBRERO', 'MARZO',
       'ABRIL', 'MAYO', 'JUNIO', 'JULIO', 'AGOSTO', 'SETIEMBRE', 'OCTUBRE',
       'NOVIEMBRE', 'DICIEMBRE', 'FILTRO']
df.columns
```

```python
df = df.dropna(how='all') #ELIMINAMOS FILAS QUE TENGAN TODOS SUS VALORES NULOS
df = df[df['OCUPACIONES'].notna()]#ELIMINAMOS FILAS QUE TENGAN VALOR NULO EN LA COLUMNA OCUPACIONES
df = df.drop(columns=['COD_OCUPACIONES' ])#eliminamos columnas irrelevantes
df=df.fillna({'FILTRO':0}) ## rellenamos la ultima columna con valores 0
df
```

```python
df.isnull().sum()#no hay nulos
```

```python
from textblob.classifiers import NaiveBayesClassifier

with open('data_entrenada.json', 'r') as fp: #### Se abre el archivo que contiene a la data entrenada
    cl=NaiveBayesClassifier(fp, format="json")  ### se le asigna la variable cl

i=0
for f in range(df.shape[0]):###iteramos sobre los indices del dataframe
    try:
        estado = cl.classify(df.OCUPACIONES[f]) ### hacemos el análisis con la data entrenada
        if (estado =="1"):      ## si el resultado es igual a (1-->pertenece a lo requerido)
            i = i + 1           ## Se adjunta un contador, para saber cuantas tipos de trabajos hay
            df.FILTRO[f]=1         ### SE LE DA EL VALOR DE 0-->1, ESTO MAS ADELANTE NOS SERVIRÁ PARA HACER EL FILTRO

    except:
        Exception
print(i)

df=df[df["FILTRO"]>0]             #### CREAMOS UN DF CON EL MISMO NOMBRE, PERO CON LA CONDICION QUE SEA MAYOR QUE 0, OSEA 1
```

```python
df=df.sort_values("OCUPACIONES",False)
df
```

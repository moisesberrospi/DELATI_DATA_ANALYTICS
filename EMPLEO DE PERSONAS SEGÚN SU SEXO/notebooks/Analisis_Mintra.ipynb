```python
train.columns =[ 'Pais','Indicador','Fuente', 'Sexo','Clasificacion', 'Año','FILTRO','status','nota','Etiqueta']
train.columns
```

```python
train
```

```python
train = train.dropna(how='all') #ELIMINAMOS FILAS QUE TENGAN TODOS SUS VALORES NULOS
train = train[train['Clasificacion'].notna()]#ELIMINAMOS FILAS QUE TENGAN VALOR NULO EN LA COLUMNA OCUPACIONES
train = train.drop(columns=['Indicador','Fuente' ,'status','nota','Etiqueta'])#eliminamos columnas irrelevantes
train
```

```python
train['FILTRO'] = 0.0
train
```

```python
train.isnull().sum()#no hay nulos
```

```python
from textblob.classifiers import NaiveBayesClassifier

with open('data_entrenada.json', 'r') as fp: #### Se abre el archivo que contiene a la data entrenada
    cl=NaiveBayesClassifier(fp, format="json")  ### se le asigna la variable cl

i=0
for f in range(train.shape[0]):###iteramos sobre los indices del dataframe
    try:
        estado = cl.classify(train.Clasificacion[f]) ### hacemos el análisis con la data entrenada
        if (estado =="1"):      ## si el resultado es igual a (1-->pertenece a lo requerido)
            i = i + 1           ## Se adjunta un contador, para saber cuantas tipos de trabajos hay
            train.FILTRO[f]=1         ### SE LE DA EL VALOR DE 0-->1, ESTO MAS ADELANTE NOS SERVIRÁ PARA HACER EL FILTRO

    except:
        Exception
print(i)

train=train[train["FILTRO"]>0]             #### CREAMOS UN DF CON EL MISMO NOMBRE, PERO CON LA CONDICION QUE SEA MAYOR QUE 0, OSEA 1
```

```python
train=train.sort_values("Clasificacion",False)
train
```

```python
train.to_csv('data_filtrada.csv', index=False) ##se exporta el dataframe a un csv
```
